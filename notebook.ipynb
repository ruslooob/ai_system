```python
import matplotlib.pyplot as plt
from core.member_functions import *
from core.operations import *

%matplotlib inline
```

```python
# объявление параметров первой ФП

# для A1
z_data = (0, 10)
z_xs = np.linspace(0, z_data[1] + 5, 100)

# для А2
trig_data = (17, 19, 20)
trig_xs = np.linspace(16, trig_data[2] + 5, 100)

# для А3
trap_data = (11, 12, 18, 21)
trap_xs = np.linspace(10, trap_data[3] + 5, 100)

# для А4
s_data = (25, 31)
s_xs = np.linspace(24, s_data[1] + 5, 100)

# для вычисления НЕ по Сугено
lmd = 0.1
```

```python
# код для задания точек для ФП Z-класса

plt.plot(z_xs, [z_member_func(x, *z_data) for x in z_xs], label='A1 original')
plt.plot(z_xs, [not_zade(z_member_func(x, *z_data)) for x in z_xs], label='NOT Zade')
plt.plot(z_xs, [not_sugeno(z_member_func(x, *z_data), lmd) for x in z_xs], label='NOT Sugeno')

plt.title('График ф-ции A1 Z-класса')
plt.xlabel('x')
plt.ylabel('μ (A1)')
plt.margins(0, 0.01)
plt.legend()
```

```python
plt.plot(s_xs, [s_member_func(x, *s_data) for x in s_xs], label='Original')
plt.plot(s_xs, [not_zade(s_member_func(x, *s_data)) for x in s_xs], label='NOT Zade')
plt.plot(s_xs, [not_sugeno(s_member_func(x, *s_data), lmd) for x in s_xs], label='NOT Sugeno')

plt.title('График ф-ции')
plt.xlabel('x')
plt.ylabel('μ')
plt.margins(0, 0.01)
plt.legend()
```

```python
plt.plot(trig_xs, [triangle_member_func(x, *trig_data) for x in trig_xs], label='Original')
plt.plot(trig_xs, [not_zade(triangle_member_func(x, *trig_data)) for x in trig_xs], label='Zade')
plt.plot(trig_xs, [not_sugeno(triangle_member_func(x, *trig_data), lmd) for x in trig_xs], label='Sugeno')

plt.title('График ФП треугольного вида и операции НЕ')
plt.xlabel('x')
plt.ylabel('μ')
plt.margins(0, 0.01)
plt.legend()
```

```python
plt.plot(trap_xs, [trapezoid_member_func(x, *trap_data) for x in trap_xs], label='Original')
plt.plot(trap_xs, [not_zade(trapezoid_member_func(x, *trap_data)) for x in trap_xs], label='NOT Zade')
plt.plot(trap_xs, [not_sugeno(trapezoid_member_func(x, *trap_data), lmd) for x in trap_xs], label='NOT Sugeno')

plt.title('График ФП трапециевидного вида и операции НЕ')
plt.xlabel('x')
plt.ylabel('μ')
plt.margins(0, 0.01)
plt.legend()
```
